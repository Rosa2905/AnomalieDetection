```python
import pandas as pd
import numpy as np
import sklearn
from sklearn.metrics import accuracy_score
from sklearn.ensemble import IsolationForest
from sklearn.model_selection import train_test_split
dt = pd.read_csv(r"C:\...................\creditcard.csv")
```

```python
x = dt.drop(['Class'], axis=1)
y = dt['Class']
```

```python
x.describe()
```

```python
import numpy as np
model = IsolationForest(
                        max_samples = 100,
                        contamination = 0.02, 
                        random_state = np.random.RandomState(60))
model = model.fit(x)
y_pred = model.predict(x)
y_pred_scores = model.decision_function(x)  #492 fraudes

y_pred = [1 if x == -1 else 0 for x in y_pred]
print("First 10 predictions: ", y_pred[:10])  
print("\n First 10 scores: ", y_pred_scores[:10])  

unique, counts = np.unique(y_pred, return_counts=True)
print("\n 0: Normal data      1: Anomalous data")
dict(zip(unique, counts))
```

```python
from sklearn import metrics
from sklearn.metrics import classification_report

print('Accuracy: {:.4f}'.format(metrics.accuracy_score(y, y_pred)))
print('Precision: {:.4f}'.format(metrics.precision_score(y, y_pred))) # No todas las positivas lo fueron
print('Recall: {:.4f}'.format(metrics.recall_score(y, y_pred)))# Encontro muchas positivas
print('F1: {:.4f}'.format(metrics.f1_score(y, y_pred)))
```

```python
## MISMO PROCEDIMIENTO PERO HACIENDO UN SPLIT DE DATA Y MinMaxScaler
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, y,train_size=0.60,test_size=0.40, random_state = 20)
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

model = IsolationForest(
                        max_samples = 100,
                        contamination = 0.17, 
                        random_state = np.random.RandomState(60))


model = model.fit(X_train)
y_pred_train = model.predict(X_train)
y_pred_train = [1 if x == -1 else 0 for x in y_pred_train]
print(" =================   Train results   =====================\n")
unique, counts = np.unique(y_pred_train, return_counts=True)
print("\n 0: Normal data      1: Anomalous data\n", "\n",dict(zip(unique, counts)))


y_pred_test = model.predict(X_test)  # these are used to calculate the performance
y_pred_test = [1 if x == -1 else 0 for x in y_pred_test]

print(" \n=================   Test results   =====================\n")
unique, counts = np.unique(y_pred_test, return_counts=True)
print("\n 0: Normal data      1: Anomalous data")
dict(zip(unique, counts))

```
